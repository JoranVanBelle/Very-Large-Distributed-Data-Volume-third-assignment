```python
def check(arr):
    sorted = True

    for i in range(len(arr)-1):
        if not arr[i]['end_date_time'] <= arr[i+1]['end_date_time']:
            print(arr[i]['end_date_time'], arr[i+1]['end_date_time'])

            sorted = False
            break
    return sorted
```
